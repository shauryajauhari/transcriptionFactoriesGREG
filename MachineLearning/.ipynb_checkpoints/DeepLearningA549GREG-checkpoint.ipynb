## <a id=link2>2. Dataset Preparation </a>

Let's check out the dataset. 

```r
a549greg <- read.table("A549forML.txt", header = TRUE) ## loading data
a549greg <- a549greg[, -c(1:3)] ## removing the first three columns for non-usage in the ML model.
```

```r
head(a549greg)
```

```r
# Classes as factor levels. 

a549greg$Class <- as.numeric(as.factor(a549greg$Class))-1
str(a549greg)
```

Great. Here **Hubs** are depicted with "0" and **Non-Hubs** by "1". As we look now, the dataset looks in close viability for application to the deep learning module. The only caveat is that the neural networks warrant the data spread between 0 and 1, for mirroring results in probability terms. Also, if chosen the sigmoid activation function shall also restrict the output in the similar range.

So if it isn't already, some sort of normalization procedure can help do that. The preferred one is the min-max normalization.

```r
# Min-Max normalization

a549greg$CTCF <- (a549greg$CTCF-min(a549greg$CTCF, na.rm = T))/(max(a549greg$CTCF, na.rm = T)-min(a549greg$CTCF, na.rm = T))
a549greg$EP300 <- (a549greg$EP300-min(a549greg$EP300, na.rm = T))/(max(a549greg$EP300, na.rm = T)-min(a549greg$EP300, na.rm = T))
a549greg$H3K27me3 <- (a549greg$H3K27me3-min(a549greg$H3K27me3, na.rm = T))/(max(a549greg$H3K27me3, na.rm = T)-min(a549greg$H3K27me3, na.rm = T))
a549greg$H3K36me3 <- (a549greg$H3K36me3-min(a549greg$H3K36me3, na.rm = T))/(max(a549greg$H3K36me3, na.rm = T)-min(a549greg$H3K36me3, na.rm = T))
a549greg$H3K4me1 <- (a549greg$H3K4me1-min(a549greg$H3K4me1, na.rm = T))/(max(a549greg$H3K4me1, na.rm = T)-min(a549greg$H3K4me1, na.rm = T))
a549greg$H3K4me2 <- (a549greg$H3K4me2-min(a549greg$H3K4me2, na.rm = T))/(max(a549greg$H3K4me2, na.rm = T)-min(a549greg$H3K4me2, na.rm = T))
a549greg$H3K4me3 <- (a549greg$H3K4me3-min(a549greg$H3K4me3, na.rm = T))/(max(a549greg$H3K4me3, na.rm = T)-min(a549greg$H3K4me3, na.rm = T))
a549greg$H3K9ac <- (a549greg$H3K9ac-min(a549greg$H3K9ac, na.rm = T))/(max(a549greg$H3K9ac, na.rm = T)-min(a549greg$H3K9ac, na.rm = T))
a549greg$H3K9me3 <- (a549greg$H3K9me3-min(a549greg$H3K9me3, na.rm = T))/(max(a549greg$H3K9me3, na.rm = T)-min(a549greg$H3K9me3, na.rm = T))
a549greg$RAD21 <- (a549greg$RAD21-min(a549greg$RAD21, na.rm = T))/(max(a549greg$RAD21, na.rm = T)-min(a549greg$RAD21, na.rm = T))
a549greg$RNAPol2 <- (a549greg$RNAPol2-min(a549greg$RNAPol2, na.rm = T))/(max(a549greg$RNAPol2, na.rm = T)-min(a549greg$RNAPol2, na.rm = T))
a549greg$YY1 <- (a549greg$YY1-min(a549greg$YY1, na.rm = T))/(max(a549greg$YY1, na.rm = T)-min(a549greg$YY1, na.rm = T))
```

Let us now formalize the dataset and bifurcate it further into training and testing cases, as per Pareto's proportions.

```r
# Data Partition

set.seed(10)
ind <- sample(2, nrow(a549greg), replace = TRUE, prob = c(0.8, 0.2))
train <- a549greg[ind==1,]
test <- a549greg[ind==2,]
```

###  <a id=link4>3.1 neuralnet </a>

```r
# Neural Networks

set.seed(12)
nn1 <- neuralnet(Class ~ .,
                data = train,
                hidden = 9,
                err.fct = "sse",
                act.fct = "logistic",
                linear.output = FALSE)
```

We can also plot the model to have a graphical visualization of the network.

```r
plot(nn1)
```

```r
# Prediction

output <- compute(nn1, train[,-13])
head(output$net.result)
head(train[1,])
```

Note, how the prediction is made. As reflected in the above graphic, we have a set of weights calculated by the model and some error terms (marked in blue).

The next step is to ascertain the overall accuracy of the model in terms of misclassification error.

```r
# Confusion Matrix & Misclassification Error - training data

p1 <- output$net.result
pred1 <- ifelse(p1>0.5, 1, 0)
tab1 <- table(pred1, train$Class)
tab1
ac <- sum(diag(tab1))/sum(tab1) #accuracy#
me <- 1 - ac
cat("The misclassification error is", me)
```

This is bizzare, though not unusual as we saw in the previous cases too. The instances of **Non-Hubs**, being larger in number are taxing the classifier to produce a bias. Truly, these numbers are insignificant. For more precise results, let us resort to the SMOTE data transformation and work again. 

```r
## installing SMOTE functionality

install.packages("DMwR")
library(DMwR)
```

```r
train$Class <- as.factor(train$Class)

set.seed(13)
smoteData <- SMOTE(Class ~., train, perc.over = 10000, k = 5, perc.under = 100)
table(smoteData$Class)
```

Let us work with this "even" data.

```r
## neural net model on SMOTE data

set.seed(14)
nn2 <- neuralnet(Class ~ .,
                data = smoteData,
                hidden = 9,
                err.fct = "sse",
                act.fct = "logistic",
                linear.output = FALSE)
```

### Note

Often times (as above), the neural network model fails to converge. To make it work, we have essentially three parameters to play with- <b>rep</b>, <b>stepmax</b>, and <b>threhold</b>.

<ul>
    <li><b><i>rep</i></b> argument is the number of times you train your neural network. Surely, higher it is, the more time it takes to process.</li>
    <li><b><i>stepmax</i></b> represents the maximum steps for training of the neural network. On attaining this threshold, the training process stops. </li>
    <li><b><i>threhold</i></b> could be increased to allow for an earlier stop of the convergence. </li>      
</ul>

These values could be tuned to successfully round-off the training process. The recommended start point is to <i>increase the <b>stepmax</b></i> value. 

```r
# Let us try again.

set.seed(15)
nn2 <- neuralnet(Class ~ .,
                 data = smoteData,
                 hidden = 9,
                 err.fct = "sse",
                 act.fct = "logistic",
                 linear.output = FALSE,
                 stepmax = 1e+09)
```

## Executing 'keras'

```r
## Loading package library.
install.packages("keras")
library(keras)

install_keras()
```

The data here gets a slightly different treatment as compared to the former package. Let us recall the structure of the data.

```r
## Let us consider our example dataset
d <- readRDS("predictionData.rds")

## Sample data for compliance with the computing infra-structure.
set.seed(108)
d <- d[sample(nrow(d), replace=F, size=0.01*nrow(d)),]

## Subsetting just features and class columns.
d <- d[,4:8]
colnames(d)[1:4] <- c("reads_h3k27ac", "reads_h3k4me1", "reads_h3k4me2", "reads_h3k4me3")


# Classes as factor levels. 
d$class <- as.numeric(as.factor(d$class))-1

str(d)

d3 <- as.matrix(d)
dimnames(d3) <- NULL
```

```r
str(d3)
```

The data was converted to a matrix form and the column names were removed.

```r
# Normalize, function is a part of keras package
d3[,1:4] <- normalize(d3[,1:4]) # normalize the features #
d3[,5] <- as.numeric(d3[,5]) # the class variable must be numeric, use of which will be known in a while #
```

```r
summary(d3)
```

The dataset is well in the ambit of the range 0 and 1. Let us now move towards preparing the training and testing data.

```r
# Data Partitioning

set.seed(1234)
ind <- sample(2, nrow(d3),replace=T, prob=c(0.7,0.3))
training <- d3[ind==1,1:4] %>% as.matrix()
trainlabels <- d3[ind==1,5] %>% as.matrix()
testing <- d3[ind==2,1:4] %>% as.matrix()
testlabels <- d3[ind==2,5] %>% as.matrix()
```

```r
# One hot encoding
traininglabels <- to_categorical(trainlabels)
testinglabels <- to_categorical(testlabels)
head(testinglabels)
```

```r
head(testlabels)
```

**One hot encoding** is a technique that transforms class labels into a proforma that the neural network shall output. The outputs from a neural network are always in terms of probabilities and the class with the highest probability is the effective output. Here we have two classes: leaf and flower that have been encoded as 0 and 1 respectively. From a binary standpoint, there could only be one class associated with each feature row, and so the other class shall be literally "OFF". So 0 is represented as [1,0] and 1 as [0,1], thereby meaning that when predicted class label is 0, it can't be 1 and vice versa.

```r
# Create sequential model
dlmodel <- keras_model_sequential()

dlmodel %>% layer_dense(units=8,
                     activation='relu',
                     input_shape=4) %>% 
            layer_dense(units=2,
                     activation='softmax')
summary(dlmodel)
```

In the above chunk of code, a sequential model type is defined. In lay terms, it means that the deep learning model will constitute of a linear array of layers. The redirectional operator (**%>%**) is deployed to channel output from one module to another. The number of input variables (features) is listed in the *input_shape* parameter, while the number of nodes in each layer is marked via *units* attribute. In addition, the activation function to be used in each node is defined via *activation* attribute. The activation function is layer specific, i.e. all nodes in a layer will use the same activation function.  
The layer_dense() function is peculiar to each layer.

```r
# Model compilation

dlmodel %>% compile (loss='binary_crossentropy',
                  optimizer='adam',
                  metrics='accuracy')
```

If the multiclass classification is sought, the *loss* attribute can take 'categorical_crossentropy' as input.

```r
# Fitting the model with training data

dlmodeltrain <- dlmodel %>% fit (training, 
                               traininglabels,
                              epoch=100, # number of iterations #
                              batch_size=10, # default=32 # 
                              validation_split=0.2) # 20% of the data #
                               
plot(dlmodeltrain)
```

There are fair indications in the graphs above. The *loss* and *accuracy* of the model aggregately follow same trend in training data and validation data. This signifies that the model is **not overfitting** and is generalizing well for newer data too.

```r
# Evaluating the model with test data

dlmodeltest <- dlmodel %>% evaluate(testing,testinglabels) 
```

```r
# Prediction and Confusion Matrix from test data

prob <- dlmodel %>% predict_proba(testing)
pred <- dlmodel %>% predict_classes(testing) %>% as.numeric()
table1 <- table(Predicted =pred, Actual= testlabels)
head(cbind(prob, pred, testinglabels))
```

```r
# Fine tuning the model

dlmodeltest
```

```r
# Changing the hidden layer units to 50.

# Create sequential model
dlmodel1 <- keras_model_sequential()

dlmodel1 %>% layer_dense(units=50,
                     activation='relu',
                     input_shape=4) %>% 
            layer_dense(units=2,
                     activation='softmax')
summary(dlmodel1)
```

```r
# Model compilation

dlmodel1 %>% compile (loss='binary_crossentropy',
                  optimizer='adam',
                  metrics='accuracy')

# Fitting the model with training data

dlmodeltrain1 <- dlmodel1 %>% fit (training, 
                               traininglabels,
                              epoch=100, # number of iterations #
                              batch_size=10, # default=32 # 
                              validation_split=0.2) # 20% of the data #
                               
plot(dlmodeltrain1)
```

The model aligns well with the training data, but is offsetted to the validation data. It seems that it overfits.

```r
# Evaluating the model with test data

dlmodeltest1 <- dlmodel1 %>% evaluate(testing,testinglabels) 

# Prediction and Confusion Matrix from test data

prob1 <- dlmodel1 %>% predict_proba(testing)
pred1 <- dlmodel1 %>% predict_classes(testing) %>% as.numeric()
table2 <- table(Predicted =pred1, Actual= testlabels)
head(cbind(prob1, pred1, testinglabels))

# Fine tuning the model

dlmodeltest1
```

There is slight difference in loss and accuracy though.

```r
# Adding a hidden layer.

# Create sequential model
dlmodel2 <- keras_model_sequential()

dlmodel2 %>% layer_dense(units=5,
                     activation='relu',
                     input_shape=4) %>% 
            layer_dense(units=50,
                     activation='relu',
                     input_shape=4) %>%
            layer_dense(units=2,
                     activation='softmax')
summary(dlmodel2)
```

```r
# Model compilation

dlmodel2 %>% compile (loss='binary_crossentropy',
                  optimizer='adam',
                  metrics='accuracy')

# Fitting the model with training data

dlmodeltrain2 <- dlmodel2 %>% fit (training, 
                               traininglabels,
                              epoch=100, # number of iterations #
                              batch_size=10, # default=32 # 
                              validation_split=0.2) # 20% of the data #
                               
plot(dlmodeltrain2)
```

```r
# Evaluating the model with test data

dlmodeltest2 <- dlmodel2 %>% evaluate(testing,testinglabels) 

# Prediction and Confusion Matrix from test data

prob2 <- dlmodel2 %>% predict_proba(testing)
pred2 <- dlmodel2 %>% predict_classes(testing) %>% as.numeric()
table3 <- table(Predicted =pred2, Actual= testlabels)
head(cbind(prob2, pred2, testinglabels))

# Fine tuning the model

dlmodeltest2
```

Infinitesimal change in loss; none in accuracy though.
